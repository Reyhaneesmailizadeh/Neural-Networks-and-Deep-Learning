```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Flatten,Dropout
import tensorflow as tf
# Checkking if GPU is available
if tf.config.list_physical_devices('GPU'):
    device_name = tf.test.gpu_device_name()
    print(f"Using GPU: {device_name}")
else:
    print("No GPU found. Using CPU.")

def create_cnn_lstm_model(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=32, kernel_size=5, activation='relu', input_shape=input_shape))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=3))
    model.add(LSTM(50, return_sequences=True, activation='tanh'))
    model.add(Dropout(0.2))
    model.add(LSTM(50, return_sequences=False, activation='tanh'))
    model.add(Dropout(0.2))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1, activation='linear'))  
    return model

input_shape = (processed_train_data.shape[1], processed_train_data.shape[2])
model = create_cnn_lstm_model(input_shape)


from tensorflow.keras.optimizers import Adam,RMSprop
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

model.compile(optimizer=RMSprop(learning_rate=0.001), loss='mean_squared_error', metrics=['mean_absolute_error'])


early_stopping = EarlyStopping(monitor='val_loss', patience=10)
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss')

history = model.fit(processed_train_data, processed_train_targets, validation_data=(processed_val_data, processed_val_targets), epochs=100, batch_size=200, callbacks=[early_stopping, model_checkpoint])

```

```python
def plot_history(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train MSE')
    plt.plot(history.history['val_loss'], label='Val MSE')
    plt.title('Model Mean Squared Error')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Squared Error')
    plt.legend(loc='upper right')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['mean_absolute_error'], label='Train MAE')
    plt.plot(history.history['val_mean_absolute_error'], label='Val MAE')
    plt.title('Model Mean Absolute Error')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Absolute Error')
    plt.legend(loc='upper right')

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

def compute_regression_metrics(true_rul,rul_pred):
    def mean_absolute_percentage_error(y_true, y_pred):
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


    # Calculate errors
    rmse = np.sqrt(mean_squared_error(true_rul, rul_pred))
    mse = mean_squared_error(true_rul, rul_pred)
    mae = mean_absolute_error(true_rul, rul_pred)
    mape = mean_absolute_percentage_error(true_rul, rul_pred)

    # Print the errors
    print("RMSE: ", rmse)
    print("MSE: ", mse)
    print("MAE: ", mae)
    print("MAPE: ", mape)

    return rmse,mse,mae,mape
```

```python
rmse,mse,mae,mape=compute_regression_metrics(true_rul,rul_pred)
print('regression CNN-LSTM with early stopping')
model_performance = pd.DataFrame({
    'Metric': ['RMSE', 'MSE', 'MAE', 'MAPE'],
    'Value': [rmse, mse, mae, mape]
})

model_performance_styled = model_performance.style.hide(axis='index').background_gradient(cmap='RdYlBu_r').format({'Value': '{:.4f}'})

model_performance_styled
```
